**Modeling**

```
from sklearn.model_selection import train_test_split
Xtrain, Xtest,Ytrain,Ytest = train_test_split(X, Y ,test_size=0.3)
```

```
Ytrain.value_counts()
```

```
DT = DecisionTreeRegressor(random_state=13579, max_depth=3, min_samples_split=5)
                         
scores = cross_val_score(DT, X, Y, cv=5, scoring='neg_mean_squared_error')
np.sqrt(-scores.mean())
```

```
KNR = KNeighborsRegressor(n_neighbors=5, weights='distance', algorithm='ball_tree',leaf_size=50)
                         
scores = cross_val_score(KNR, X, Y, cv=5, scoring='neg_mean_squared_error')
np.sqrt(-scores.mean())
```

```
SV = SVR(kernel='rbf', degree=3,gamma='auto_deprecated',coef0=0.0,tol=0.001,
        shrinking=True,cache_size=400,verbose=False,max_iter=1)
scores = cross_val_score(KNR, X, Y, cv=5, scoring='neg_mean_squared_error')
np.sqrt(-scores.mean())
```

```
LR = LinearRegression(fit_intercept=True,normalize=False,copy_X=True,n_jobs=None)
scores = cross_val_score(KNR, X, Y, cv=5, scoring='neg_mean_squared_error')
np.sqrt(-scores.mean())
```

```
MLP = MLPRegressor(random_state=13579, learning_rate_init=0.01, hidden_layer_sizes=(50,20,20))
scores = cross_val_score(MLP, X, Y, cv=5, scoring='neg_mean_squared_error')
np.sqrt(-scores.mean())
```
